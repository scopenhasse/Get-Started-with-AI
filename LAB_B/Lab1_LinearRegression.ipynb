# Linear Regression

Let's fabricate some data that shows a relationship between page speed and amount purchased:

```python
%matplotlib inline
import numpy as np
from pylab import *

pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = 100 - (pageSpeeds + np.random.normal(0, 0.1, 1000)) * 3

scatter(pageSpeeds, purchaseAmount)

# the amount of purchase is getting higher as the page loading speed is getting higher
```

```python
# inverse the scatter plot
pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = 100 + (pageSpeeds + np.random.normal(0, 0.1, 1000)) * 3

scatter(pageSpeeds, purchaseAmount)
```

As we only have two features, we can keep it simple and just use scipy.state.linregress:

```python
from scipy import stats

slope, intercept, r_value, _ , std_err = stats.linregress(pageSpeeds, purchaseAmount)

# stats.linregress returns the slope, intercept, r_value, p_value, std_err of the linear regression
# slope means the spill or flow of the line
# intercept means the intercept of the line
# r_value means the correlation coefficient
# p_value means the p_value of the linear regression
# std_err means the standard error of the linear regression
```

The R-squared value shows the following fit:

```python
r_value ** 2
# r_value ** 2 means the coefficient of determination of the linear regression
# the coefficient of determination is the proportion of the variance in 
# the dependent variable that is predictable from the independent variable
```

Let's use the slope and intercept we got from the regression to plot predicted values vs. observed:

```python
import matplotlib.pyplot as plt

def predict(x):
    return slope * x + intercept

fitLine = predict(pageSpeeds)

plt.scatter(pageSpeeds, purchaseAmount)
plt.plot(pageSpeeds, fitLine, c='r')
plt.show()

# the red line is the linear regression line
# the predict function is used to predict the purchase amount according to the page loading speed
# the predict function is the linear regression function
```

## Activity

Try increasing the random variation in the test data, and see what effect it has on the r-squared error value.

```python
import numpy as np
from pylab import *

pageSpeeds = np.random.normal(3.0, 5, 1000)
purchaseAmount = 100 - (pageSpeeds + np.random.normal(0, 0.5, 1000)) * 3

slope, intercept, r_value, _ , std_err = stats.linregress(pageSpeeds, purchaseAmount)


scatter(pageSpeeds, purchaseAmount)

print("R-squared =",r_value ** 2)

def predict(x):
    return slope * x + intercept


fitLine = predict(pageSpeeds)

plt.scatter(pageSpeeds, purchaseAmount)
plt.plot(pageSpeeds, fitLine, c='g')
plt.show()
```

```python
#the R squared value is 0.98, which means that the model is very good, and the data is very close to the fitted line
```

```python
import numpy as np
from pylab import *

pageSpeeds = np.random.normal(3.0, 0.1, 1000)
purchaseAmount = 100 - (pageSpeeds + np.random.normal(0, 0.5, 1000)) * 3

slope, intercept, r_value, _ , std_err = stats.linregress(pageSpeeds, purchaseAmount)

scatter(pageSpeeds, purchaseAmount)

print("R-squared =",r_value ** 2)

def predict(x):
    return slope * x + intercept


fitLine = predict(pageSpeeds)

plt.scatter(pageSpeeds, purchaseAmount)
plt.plot(pageSpeeds, fitLine, c='g')
plt.show()
```

```python
# R squared value is 0.053 which means that the model is not good, and the data is not close to the fitted line

# lets plot R squared value for different standard deviation of the page loading speed
# the standard deviation of the page loading speed is the variance of the page loading speed
# the variance of the page loading speed is the amount of the page loading speed is changing


x= np.arange(0.1, 5, 0.1)
R_values = []

for i in x :
    pageSpeeds = np.random.normal(3.0, i, 1000)
    purchaseAmount = 100 - (pageSpeeds + np.random.normal(0, 0.5, 1000)) * 3

    slope, intercept, r_value, _ , std_err = stats.linregress(pageSpeeds, purchaseAmount)

    R_values.append(r_value ** 2)

scatter(x, R_values)
xlabel('Standard Deviation of the Page Loading Speed')
ylabel('R squared value')

# the R squared value is getting higher as the standard deviation of the page loading speed is getting lower
# the R squared value is getting lower as the standard deviation of the page loading speed is getting higher

```
