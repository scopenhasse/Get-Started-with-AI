# Exercise: Mean & Median Customer Spend

Here's some code that will generate some random e-commerce data; just an array of total amount spent per transaction. Select the code block, and hit "play" to execute it:

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
#.normal : makes the dirtubution of the random numbers follows the normal law gaussian law 
incomes = np.random.normal(100.0, 20.0, 10000)
print(incomes)
plt.hist(incomes, 50)
plt.show()
```

Now, find the mean and median of this data. In the code block below, write your code, and see if your result makes sense:

```python
# from the histogram we expect to get the result close to 100 
print("MEAN:")
print(np.mean(incomes))
print("MEDIAN:")
print(np.median(incomes))
```

This is pretty much the world's easiest assignment, but we're just trying to get your hands on iPython and writing code with numpy to get you comfortable with it.

Try playing with the code above to generate different distributions of data, or add outliers to it to see their effect.

```python
#lets generate mmore random data : 
Data = np.random.normal(10000, 20.0, 10000)
print(Data)
plt.hist(Data, 50)
plt.show()
```

```python
#Okey we expecting to get the mean and the median around 10000 lets see : 
print("MEAN:")
print(np.mean(Data))
print("MEDIAN:")
print(np.median(Data))
```

```python
#lets add more outliers
Data = np.append(Data, [1000000000])
Data = np.append(Data, [200])

#this plot here is to show that outliers might ruin an analysis
plt.hist(Data, 50)
plt.show()
```

```python
print("MEAN:")
print(np.mean(Data))
print("MEDIAN:")
print(np.median(Data))
```
